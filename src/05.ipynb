# 演習：データ分析を行い、データに基づく営業支援のための施策を立てる
ここまでの分析で、下記のことがわかりました。

- すでにローンを持っていたり、起業家,自営業などの職業の人が成約後、預金金額（売上の）多い傾向にあることがわかった。
- 起業家、自営業の成約率は高くない
- Retired は成約率は高いが売上はあまり上がらない（成約してもそこまで預金しない）ことがわかった。
- 家政婦や無職の人は成約を行っても預金が少ない（売上が低い）ことがわかった。

更に分析を進めて、最終的な結論と具体的な営業支援の施策を考えましょう。  

現状での施策を考えることはできますが、単に企業やすでにローンを持っている人と行っても様々なタイプの方がいます。より具体的で効果性の高い、施策となるようにさらなる分析を行いましょう。
