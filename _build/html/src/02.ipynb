# 問題の把握・明確化

ここでは、データの収集と整理を行い、収集されたデータの理解を深めます。グラフ化などは一部、分析プロセスデモ実施しますが、ここではデータの理解を目的として使用できます。

## 概要

データ分析を効果的に進めるには、まず「何を解くべきか」を明確にすることが重要です。  
このフェーズでは、データの収集・整理を行いながら、収集されたデータへの理解を深めます。  
グラフ化やピボットテーブルは一部ここでも活用しますが、主な目的は**データを理解すること**です。

---

## 効率の良いデータ収集

実務では、分析開始時点でデータが揃っていることは少ないです。  
そのため、**集めやすいデータから優先して収集する**考え方が重要になります。

### データの種類

| 種類 | 説明 | 例 | 注意点 |
|------|------|----|--------|
| **一次データ** | 目的に合わせて新たに収集するデータ | アンケート、インタビュー | 時間・コストがかかる |
| **二次データ** | 他の目的で既に収集済みのデータ | 販売データ、顧客データ、官公庁の統計 | 必ずしも目的に適していない。加工や個人情報の取り扱いに注意 |

### 収集の進め方

1. **イシューツリーを活用**して、大枠の問題を頂点に、要因・仮説・必要なデータを展開する
2. データが絞り込まれたら、**二次データのように素早く集められるものを優先**する
3. データソースは**社内（内部）だけでなく、社外（外部）も候補**に入れる
4. 全てのデータが揃わなくても、**集まったデータで仮説の検証を開始**する

> ポイント：データを闇雲に収集するのではなく、仮説ドリブンで必要なデータを特定する。

---

## データの最適化

二次データは分析目的で集められたものではないため、**関係のないデータが混在**しています。  
分析に入る前にデータを取捨選択・最適化する必要があります。

### ポイント① 必要最低限よりも幅をもたせる

- 分析・検証に必要なデータに絞り込みすぎない
- 分析が進む中で、**異なる切り口が必要になることがある**

**例：** 商品Aを分析する場合  
→ Aの販売データだけでなく、同じ系列の商品や競合商品のデータも含めると、より多角的な検証が可能になる

### ポイント② 必要に応じて異常値を取り除く

明らかに他のデータと乖離した値（異常値）は分析結果に悪影響を与える可能性があります。

**異常値の主な原因**
- 自然災害やインシデントなど突発的なイベント
- 入力ミス・システム障害によるデータ誤り

**異常値の検出・除去方法**
- 散布図にプロットして視覚的に確認する
- **3シグマ法**（平均 ± 3σ の範囲外を異常値とみなす）を使う

> 注意：異常値そのものが分析対象になるケースもあるため、取り除くかどうかは状況に応じて判断する。

---

## データの理解を深める

データの取捨選択と並行して、**収集したデータが何を示しているかを理解する**ことが重要です。

### 理解を深める手法

| 手法 | 目的 |
|------|------|
| グラフ化（棒グラフ・折れ線・散布図など） | データの傾向・分布・異常値を視覚的に把握する |
| ピボットテーブル | データを集計・クロス分析して構造を理解する |
| 基本統計量（平均・中央値・標準偏差など） | データの全体像を数値で把握する |

- データを深く理解することで、**新たな仮説が生まれること**もある
- 分析・検証はデータを理解しながら進める反復プロセス

---

## まとめ

```
問題の把握・明確化フェーズのステップ

① イシューツリーで問題と仮説を構造化する
        ↓
② 二次データを優先しながらデータを収集する（内部・外部）
        ↓
③ 分析目的に合わせてデータを取捨選択・最適化する
        ↓
④ グラフやピボットテーブルでデータを理解し、新たな仮説を立てる
        ↓
⑤ 仮説の検証フェーズへ進む
```

このフェーズを丁寧に行うことで、その後の分析の精度と効率が大きく向上します。



# データの収集

ここでは、データの収集と整理を行い、収集されたデータの理解を深めます。グラフ化などは一部、分析プロセスデモ実施しますが、ここではデータの理解を目的として使用できます。

---

## 効率の良いデータ収集
本講座ではデモデータを用意していますが、実務においてはじめからデータが用意されていることは多くありません。データ収集の考え方として、集めやすいデータから集める方法があります。データは大きく一次データと二次データに分けられます。

- 一次データ: 目的に沿って新たに収集されるデータ。収集には時間とコストがかかります。例えば、分析の目的に合わせたアンケートデータなどです。

- 二次データ: 他の目的で既に収集されたデータ。具体的には、通常業務で収集された販売データや顧客データ、官公庁などの統計データ。必ずしも目的に適したデータではないため、加工や個人情報を含むデータの取り扱いには注意が必要です。 

データは闇雲に収集してはいけません。先ほど紹介したイシューツリーを用いて、大枠の問題を頂点として、考えられる要因や仮説、それらを検証するデータを展開します。そして、データが絞り込まれてきたら、二次データのようにすぐに収集できるデータを優先して集めます。データソースは内部だけでなく、外部のデータも候補に入れます。全てのデータが揃わなくても、集まったデータで仮説の検証を始めます。

---

## データの最適化
二次データを収集し分析しようとすると、そもそもデータが分析の目的で集められたものではないため、関係のないデータが混在しています。このようなデータは取捨選択する必要があります。

ポイント

- 必要最低限必要なデータよりも余裕をもたせる

検証・分析に必要なデータに絞り込むのではなく、幅を持たせます。分析や検証を行っていく中で、異なる切り口で分析をすることがあります。例えば、ある商品Aを分析する際、Aの販売データだけでなく、同じ系列や競合のデータを含めることで、より多角的な検証が可能になります。

- 必要に応じて異常値を取り除く

明らかに他のデータより値が大きい場合などは、異常値として取り除く必要があります。異常値は分析や検証においてマイナスの要因となる可能性があります。例えば、自然災害により集客ができなかった場合などの突然発生するインシデントによる異常値や、入力ミスによる異常値が含まれることがあります。異常値の分析が必要な場合もありますので、適宜判断が求められます。散布図にプロットすることで異常値を見つけやすくなります。また、外れ値除去の方法として3シグマ法などがあります・

---

## データの理解をする
データの取捨選択と並行して、データの理解を進めます。収集したデータがどのようなものかわからなければ、分析や検証を行うことはできません。データは必要に応じてグラフ化したり、ピボットテーブルを作成することで、より理解を深めることができます。結果によっては新たな仮説を立てることもあります。

それでは、本講座で使用する銀行のデータを見てみましょう。
