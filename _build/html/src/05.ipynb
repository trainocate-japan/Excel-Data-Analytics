# 【演習】データ分析を行い、データに基づく営業支援のための施策を立てる

ここまでの分析で、下記のことがわかりました。

【単一のデータの可視化による気づき】


- リストの平均年齢41歳
- リストの平均残高1450.296ドル
- 成約数、不成約の数はあまり変わらない
- 2000 サンプルと16列あることがわかる
- ローンのない顧客が大半を占める
- 顧客のボリュームゾーンは 40 代　

【複数データの集計・可視化による気づき】

- 「工場作業員」や「接客」の仕事に就いている顧客は、断られることが多い
- 既婚の顧客は成約率が低い傾向にあるがデータ数は一番多い。
- 最終学歴が中等教育だと成約率が低い。学歴が上がるほど成約率が上がっている。
- 学生の成約率も高い
- 工場作業員・接客業・家政婦は成約率が低い。だが現状の顧客リストに工場作業員は 2 番目に多くいると考えられる

分析・検定

- 売上には銀行残高が大きく関わることがわかった
- 預金口座の成約の有無にはローンの有無と持ち家の有無の関わりがあることがわかった
- 登録されている顧客への連絡手段も成約に関わりがあることがわかった
- 成約と有無と結婚状況には関連がある



更に分析を進めて、最終的な結論と具体的な営業支援の施策を考えましょう。  

現状での施策を考えることはできますが、単に家やすでにローンを持っている人と行っても様々なタイプの方がいます。ここまでの分析結果をもとに施策を考えることもできますが、より具体的で効果の高い施策となるように更なる分析を行いましょう。
