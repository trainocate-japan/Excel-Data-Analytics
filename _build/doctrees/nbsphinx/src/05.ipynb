# 総演習：データに基づく営業支援のための施策を立案

ここまでの分析で、下記のことがわかりました。

【単一のデータの可視化による気づき】


- リストの平均年齢41歳
- リストの平均残高1790ドル
- 成約確度は3,7が多い
- ローンのない顧客が大半を占める
- 顧客のボリュームゾーンは 40 代　

【複数データの集計・可視化による気づき】

- 「工場作業員」や「接客」の仕事に就いている顧客は、断られることが多い
- 成約数の最も多いのはマネージャー
- データ数に対して引退の成約確度が高そう

【分析・検定】

- すべての値が成約確度に影響を与えている。結婚状況と持ち家の有無の影響度合いが高い
- 成約確度と結婚状況には統計的に関連がある


更に分析を進めて、最終的な結論と具体的な営業支援の施策を考えましょう。  

現状での施策を考えることはできますが、単に家やすでにローンを持っている人と行っても様々なタイプの方がいますより具体的で効果の高い施策となるように更なる分析を行いましょう。
